```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
import xgboost as xgb  # make sure xgboost is installed
import joblib
import streamlit as st
```

```python
# Load the dataset
df = pd.read_csv("Salary Data.csv")
```

```python
# Display basic information
print("Initial data info:")
df.info()
```

```python
# Show the first few rows
print("\nSample data:")
df.head()
```

```python
# Check for missing values
print("\nMissing values before cleaning:")
print(df.isnull().sum())
```

```python
# Display rows with missing values
print("\nRows with missing data:")
print(df[df.isnull().any(axis=1)])
```

```python
# Drop rows where all values are missing
df_cleaned = df.dropna()
```

```python
# Confirm no missing values remain
print("\nShape after dropping missing rows:", df_cleaned.shape)
print("\nMissing values after cleaning:")
print(df_cleaned.isnull().sum())
```

```python
# Separate features and target
X = df_cleaned.drop("Salary", axis=1)
y = df_cleaned["Salary"]
```

```python
# Identify categorical columns
categorical_cols = ["Gender", "Education Level", "Job Title"]
```

```python
# Create ColumnTransformer to encode categorical columns
column_transformer = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(drop="first"), categorical_cols)
    ],
    remainder="passthrough"  # Keep numerical columns as they are
)
```

```python
# Fit and transform the features
X_encoded = column_transformer.fit_transform(X)
```

```python
# Get the OneHotEncoder instance
ohe = column_transformer.named_transformers_["cat"]
```

```python
# Get feature names for encoded columns
encoded_feature_names = ohe.get_feature_names(categorical_cols)
```

```python
# Combine with numerical column names (those that were 'passed through')
numeric_cols = [col for col in X.columns if col not in categorical_cols]
all_feature_names = list(encoded_feature_names) + numeric_cols
```

```python
# Create DataFrame
X_encoded_df = pd.DataFrame(X_encoded.toarray(), columns=all_feature_names)
X_encoded_df.head()
```

```python
# Split data (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
# Train Linear Regression model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)
```

```python
# Predict on test set
y_pred = lr_model.predict(X_test)
```

```python
# Evaluation metrics
print(" Linear Regression Results:")
print("R Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
print("MSE:", mean_squared_error(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# Store results
results = {}
```

```python
# 1 Decision Tree
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)
dt_pred = dt_model.predict(X_test)
results["Decision Tree"] = {
    "R2": r2_score(y_test, dt_pred),
    "MAE": mean_absolute_error(y_test, dt_pred),
    "RMSE": np.sqrt(mean_squared_error(y_test, dt_pred))
}
```

```python
# 2 Random Forest
rf_model = RandomForestRegressor(random_state=42)
rf_model.fit(X_train, y_train)
rf_pred = rf_model.predict(X_test)
results["Random Forest"] = {
    "R2": r2_score(y_test, rf_pred),
    "MAE": mean_absolute_error(y_test, rf_pred),
    "RMSE": np.sqrt(mean_squared_error(y_test, rf_pred))
}
```

```python
# 3 XGBoost
xgb_model = xgb.XGBRegressor(random_state=42)
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
results["XGBoost"] = {
    "R2": r2_score(y_test, xgb_pred),
    "MAE": mean_absolute_error(y_test, xgb_pred),
    "RMSE": np.sqrt(mean_squared_error(y_test, xgb_pred))
}
```

```python
#  Display results
for model_name, metrics in results.items():
    print(f"\n {model_name} Performance:")
    for metric_name, value in metrics.items():
        print(f"{metric_name}: {value:.4f}")
```
